```python
class Node:
    def __init__(self, data):
        self.data = data
        self.h = None
        self.n = None
class DoubleLL:
    def __init__(self):
        self.head = None
    def iab(self, data):
        nn = Node(data)
        nn.n = self.head
        if self.head:
            self.head.h = nn
        self.head = nn
    def iae(self, data):
        nn = Node(data)
        if self.head is None:
            self.head = nn
            return
        current = self.head
        while current.n:
            current = current.n
        current.n = nn
        nn.h = current
    def iap(self, data, p):
        nn = Node(data)
        if self.head is None:
            print("it is an empty list, there are no positions for the insertion")
            return
        current = self.head
        c = 1
        while c!=(p-1):
            current = current.n
            if current is None:
                print("Out of The Given Range")
                print("The total Elements are",c)
            c += 1
        nn.h = current
        nn.n = current.n
        current.n.h = nn
        current.n = nn
    def search(self, data):
        if self.head is None:
            print("An Empty List No Elements are Present")
            return
        current = self.head
        c = 1
        while current.data != data:
            current = current.n
            c += 1
            if current is None:
                print(f"{data} Element Not Found in the DLL")
                return
        print(f"{data} element found at pos {c}")
    def delete(self, data):
        if self.head is None:
            print("An Empty List No Elements are Present")
            return
        if self.head.data == data:
            self.head.n.h = self.head.h
            self.head = self.head.n
            print(f"{data} Deleted Successfully at head")
            return
        current = self.head
        while current.data != data:
            current = current.n
            if current is None:
                print(f"{data} Element Not Found in the DLL")
                return
        if current.n is None:
            current.h.n = current.n
            print(f"{data} Deleted Successfully at End")
            return
        current.h.n = current.n
        current.n.h = current.h
        print(f"{data} Deleted Successfully")
    def backtraverse(self):
        if self.head is None:
            print("An Empty DLList")
            return
        current = self.head
        while current.n:
            current = current.n
        while current:
            print(f"{current.data}",end = "<-->")
            current = current.h
        print("None")

    def dab(self, rep=1):
        if self.head is None:
                print("No Elements in the DLL")
                return
        for _ in range(rep):
            if self.head is None:
                print("No Elements left in the DLL")
                return
            self.head = self.head.n
            if self.head:
                self.head.h = None
        
    def dae(self, rep=1):
        for _ in range(rep):
            if self.head is None:
                print("No Elements there in the DLL")
                return
            current = self.head
            while current.n:
                current = current.n
            if not current.h:
                self.head = None
                print("No Elements Left to be deleted")
                return
            current.h.n = None
            print("Deleted successfully")

    def display(self):
        if self.head is None:
            print("AN EMPTY DLL")
            return
        current = self.head
        while current:
            print(f"{current.data}",end = "<-->")
            current = current.n
        print("None")

lis = list(map(int, input().split()))
dll = DoubleLL()
print("How do you want to insert the elements: ")
print("1.IAB\n2.IAE")
while True:
    chc = int(input("Enter type of Insertion option: "))
    if chc == 1:
        for _ in lis:
            dll.iab(_)
        break
    elif chc == 2:
        for _ in lis:
            dll.iae(_)
        break
    else:
        print("choose from above two options only")
print("<--- DOUBLE LINKED LIST --->")
while True:
    print("1.IAB\n2.IAE\n3.IAP\n4.SEARCH\n5.DELETE\n6.DISPLAY\n7.BACK_TRAVERSES\n8.DAB\n9.DAE")
    ch = int(input("Enter ur choice : "))
    if ch==1:
        data = int(input("Enter the data : "))
        dll.iab(data)
    elif ch == 2:
        data = int(input("Enter the data : "))
        dll.iae(data)
    elif ch == 3:
        data = int(input("Enter the data : "))
        p = int(input("Enter the position : "))
        dll.iap(data, p)
    elif ch == 4:
        data = int(input("Enter the data : "))
        dll.search(data)
    elif ch == 5:
        data = int(input("Enter the data : "))
        dll.delete(data)
    elif ch == 6:
        dll.display()
    elif ch == 7:
        dll.backtraverse()
    elif ch == 8:
        r = int(input("how many times to do dab : "))
        dll.dab(r)
    elif ch == 9:
        r = int(input("how many times to do dae : "))
        dll.dae(r)
    else:
        break
```

```python
def dab(self):
    if self.head is None:
        print("No Elements in the DLL")
        return
    self.head = self.head.n
    self.head.h = None
def dae(self):
    if self.head is None:
        print("No Elements in the DLL")
        return
    current = self.head
    while current.n:
        current = current.n
    current.h.n = None
```
